```python
#Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:Brand,ProductDescription,Price.

import requests
from bs4 import BeautifulSoup

def scrape_sunglasses_data(url):
  response = requests.get(url)
  soup = BeautifulSoup(response.content, 'html.parser')
  
  sunglasses = soup.find_all('div', {'class': '_2WkVRV'})
  
  data = []
for sunglass in sunglasses:
        brand = sunglass.find('div',class_=="_2WkVRV")
        description = sunglass.find('a',class_=="IRpwTa")
        price = sunglass.find('div',class_=="_30jeq3")
  
data.append({'Brand': brand, 'ProductDescription': description, 'Price': price})
  
return data

all_data = []
page_url = "https://www.flipkart.com/"
count = 0

while count < 100:
  data = scrape_sunglasses_data(page_url)
  all_data.extend(data)
  
  next_button = soup.find('a', {'class': '_1LKTO3'})
  if next_button:
    page_url = "https://www.flipkart.com/" + next_button['href']
  else:
    break
  
  count += len(data)
```
